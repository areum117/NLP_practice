# BoW 기반 텍스트 분석 연습

1. 적절한 데이터셋을 찾거나 생성하고, 적절한 전처리를 진행한다. [01_preprocessing.ipynb]
2. TF-IDF Vectorizer를 이용하여 벡터화한다.
3. Cosine Similarity를 계산하여 입력된 문자열의 긍/부정을 판단한다.

```python
import pandas as pd
```

```python
text_df = pd.read_csv('./preprocessing_data.csv', encoding='utf-8')
text_df
```

```python
import ast

text_df['tokens'] = text_df['tokens'].apply(ast.literal_eval)
text_df['string'] = text_df['tokens'].apply(lambda x: ' '.join(x))

text_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
text_vecs = tfidf_vectorizer.fit_transform(text_df['string'])

text_vecs
```
